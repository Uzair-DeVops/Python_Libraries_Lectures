```python

# Load the image
image = imageio.imread('image.jpg')

# Convert the image to grayscale
gray_image = np.dot(image[...,:3], [0.2989, 0.5870, 0.1140])

# Create a gray color overlay
gray_overlay = np.zeros_like(image)  # Create an array of zeros with the same shape as the original image
gray_overlay[..., 0] = gray_image  # Assign gray image to the red channel
gray_overlay[..., 1] = gray_image  # Assign gray image to the green channel
gray_overlay[..., 2] = gray_image  # Assign gray image to the blue channel

# Blend the original image with the gray overlay
alpha = 0.5  # Adjust the transparency of the overlay
blended_image = (1 - alpha) * image + alpha * gray_overlay

# Ensure the values are in the valid range (0-255 for image)
blended_image = np.clip(blended_image, 0, 255).astype(np.uint8)

# Display the blended image
plt.imshow(blended_image)
plt.axis('off')  # Hide the axes
plt.show()
```

```python
import numpy as np
import imageio
import matplotlib.pyplot as plt

# Load the image
image = imageio.imread('image.jpg')  # Replace 'image.jpg' with your image file path

# Check the shape of the image
height, width, channels = image.shape
print(f"Image shape: {height} (height) x {width} (width) x {channels} (channels)")

# Create an orange image by manipulating the RGB channels
orange_image = np.zeros_like(image)  # Create an empty array with the same shape as the original image

# Set the Red and Green channels to create an orange tint
orange_image[..., 0] = image[..., 0]  # Keep the Red channel
orange_image[..., 1] = (image[..., 1] * 0.5)  # Reduce the Green channel to keep a balance
orange_image[..., 2] = 0  # Set the Blue channel to zero

# Display the original and orange images
plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

# Orange Image
plt.subplot(1, 2, 2)
plt.imshow(orange_image)
plt.title("Orange Image")
plt.axis("off")

plt.show()
```

```python
import numpy as np
import imageio
import matplotlib.pyplot as plt

# Load the image
image = imageio.imread('image.jpg')  # Replace 'image.jpg' with your image file path

# Check the shape of the image
height, width, channels = image.shape
print(f"Image shape: {height} (height) x {width} (width) x {channels} (channels)")

# Create a golden image by manipulating the RGB channels
golden_image = np.zeros_like(image)  # Create an empty array with the same shape as the original image

# Set the Red and Green channels to create a golden tint
golden_image[..., 0] = image[..., 0] * 1.5  # Increase the Red channel intensity
golden_image[..., 1] = image[..., 1] * 1.2  # Increase the Green channel intensity
golden_image[..., 2] = image[..., 2] * 0.5  # Reduce the Blue channel intensity

# Clip values to be in the range [0, 255] to avoid overflow
golden_image = np.clip(golden_image, 0, 255).astype(np.uint8)

# Display the original and golden images
plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

# Golden Image
plt.subplot(1, 2, 2)
plt.imshow(golden_image)
plt.title("Golden Image")
plt.axis("off")

plt.show()
```

```python
import numpy as np
import imageio
import matplotlib.pyplot as plt

# Load the image
image = imageio.imread('image.jpg')  # Replace 'image.jpg' with your image file path

# Check the shape of the image
height, width, channels = image.shape
print(f"Image shape: {height} (height) x {width} (width) x {channels} (channels)")

# Create a gray image by manipulating the RGB channels
gray_image = np.zeros_like(image)  # Create an empty array with the same shape as the original image

# Set all channels to the average value of the original RGB channels to create gray
gray_image[..., 0] = image[..., 0] * 0.2989 + image[..., 1] * 0.5870 + image[..., 2] * 0.1140  # Red channel
gray_image[..., 1] = gray_image[..., 0]  # Green channel (same as red)
gray_image[..., 2] = gray_image[..., 0]  # Blue channel (same as red)

# Clip values to be in the range [0, 255] to avoid overflow
gray_image = np.clip(gray_image, 0, 255).astype(np.uint8)

# Display the original and gray images
plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

# Gray Image
plt.subplot(1, 2, 2)
plt.imshow(gray_image)
plt.title("Gray Image")
plt.axis("off")

plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# Load the image
image = Image.open('image.jpg')  # Replace 'image.jpg' with your image file path

# Convert the image to RGB (if it's not in that mode)
image = image.convert('RGB')

# Convert the image to an array
image_array = np.array(image)

# Apply a purple tint by enhancing red and blue channels
purple_tinted_image_array = image_array.copy()
purple_tinted_image_array[..., 1] = 0  # Set green channel to 0

# Optionally, enhance the red and blue channels (for example, by multiplying them)
purple_tinted_image_array[..., 0] = purple_tinted_image_array[..., 0] * 0.7  # Reduce red intensity
purple_tinted_image_array[..., 2] = purple_tinted_image_array[..., 2] * 0.7  # Reduce blue intensity

# Convert back to image
purple_tinted_image = Image.fromarray(np.uint8(purple_tinted_image_array))

# Display the original and purple-tinted images
plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

# Purple-Tinted Image
plt.subplot(1, 2, 2)
plt.imshow(purple_tinted_image)
plt.title("Purple-Tinted Image")
plt.axis("off")

plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# Load the image
image = Image.open('image.jpg')  # Replace 'image.jpg' with your image file path

# Convert the image to RGB (if it's not in that mode)
image = image.convert('RGB')

# Convert the image to an array
image_array = np.array(image)

# Apply a blue tint by increasing the blue channel
# This can be done by setting red and green to 0, or enhancing blue channel
blue_tinted_image_array = image_array.copy()
blue_tinted_image_array[..., 0] = 0  # Set red channel to 0
blue_tinted_image_array[..., 1] = 0  # Set green channel to 0

# Convert back to image
blue_tinted_image = Image.fromarray(blue_tinted_image_array)

# Display the original and blue-tinted images
plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

# Blue-Tinted Image
plt.subplot(1, 2, 2)
plt.imshow(blue_tinted_image)
plt.title("Blue-Tinted Image")
plt.axis("off")

plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt

# Load the image
image = Image.open('image.jpg')  # Replace 'image.jpg' with your image file path

# Convert to different color modes
gray_image = image.convert('L')    # Grayscale
binary_image = image.convert('1')  # Black and White
cmyk_image = image.convert('CMYK') # CMYK mode

# Display the images
plt.figure(figsize=(12, 4))

# Grayscale Image
plt.subplot(1, 3, 1)
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")

# Binary Image
plt.subplot(1, 3, 2)
plt.imshow(binary_image, cmap='gray')
plt.title("Binary Image")
plt.axis("off")

# CMYK Image (displaying it as RGB for visualization)
plt.subplot(1, 3, 3)
plt.imshow(cmyk_image.convert('RGB'))  # Convert CMYK back to RGB for display
plt.title("CMYK Image")
plt.axis("off")

plt.show()
```
