# Duplicated funs
it is used to find a duplicate items it have one parameter (keep ) if we put the value of keep = "first" it will put the first value to false and then other true and vise versa and if i put keeo = false if will put all the duplicate value to true 

```python
df.sort_values("First Name" , inplace=True)
```

```python
df["First Name"]
```

```python
df["First Name"].duplicated(keep="first")
```

```python
df["First Name"]
```

```python
df["First Name"].duplicated(keep="last")
```

```python
df["First Name"].duplicated(keep=False)
```

# (~)  this sign will change our true to false 

```python
df[~df["First Name"].duplicated(keep=False)]
```

# drop_duplicate 
it works on df not series

```python
df.drop_duplicates(subset=["First Name"] , keep=False)
```
